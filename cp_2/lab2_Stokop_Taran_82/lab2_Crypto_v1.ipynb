Частоты букв в тексте:

```python
def freq_func(filename):
    with open(filename) as f:
        c = Counter()
        for line in f:
            c += Counter(line)
    freq = dict(c)
    return freq
```

```python
freq_func('clearText')
```

```python
def accordance_index(some_dict, number):
    result_index=0
    for i in some_dict:
        result_index+=some_dict[i]*(some_dict[i]-1)
    result_index=result_index/(len(number)*(len(number)-1))
    return result_index
```

```python
def f1(number_block,some_encrypt_text,r):
    temp_list=[]
    for i in range(number_block,len(some_encrypt_text),r):
        temp_list.append(some_encrypt_text[i])
            
    amount_lett_every_block = collections.Counter(temp_list) #amount of letters for every block
    
    symbol_max= max(amount_lett_every_block, key=amount_lett_every_block.get)  
    
    y_with_star=int(our_alphabet[symbol_max]) #переводим букву в число у*
    k_key=(y_with_star-14)%32                 #вычисляем k, 14 - значение наиболее расспостраненной буквы русс. ал - "о"
    
    
    key.append(our_alp[k_key])
    return print('блокY=',number_block,': наибольшая частота у буквы: ',symbol_max,' => Буква ключа:', our_alp[k_key],'\n')
        
    
key=[]
for k in range(0,len_our_block):
    f1(k,encrypt_text,len_our_block)   
```
