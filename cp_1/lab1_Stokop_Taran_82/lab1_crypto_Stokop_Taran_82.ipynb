```python
import codecs as cs
import re
from collections import Counter
import math
```

Фільтрація текстового файлу

```python
def clear_text(file, withSpaces):
        text = cs.open(file,encoding='utf-8')
        text = text.read().lower()
        text = text.replace('ё','е')
        text = text.replace('ъ','ь')
        text = re.sub('[^а-я ]', '', text) if withSpaces else re.sub('[^а-я]', '', text)
        fileName = "textWithSpaces" if withSpaces else "textWithoutSpaces"
        open(fileName,'w', encoding='utf-8').write(text)
```

```python
clear_text('prestuplenie.txt', False)
```

Частота монограми

```python
def mono_func(filename):
    with open(filename, encoding='utf-8') as f:
        c = Counter()
        for line in f:
            c += Counter(line)
    with open(filename, 'r', encoding='utf-8') as file:
        data = file.read()
        
    total = len(data)
    freq_mono = dict(c)
    
    for k in freq_mono:
        freq_mono[k] = float(freq_mono[k]/total)
    return dict(sorted(freq_mono.items(), key=lambda x: x[1], reverse=True))
```

```python
mono_func('textWithSpaces')
```

Частоти для біграмм


```python
def bigram_func(filename, inter):
    with open(filename, 'r', encoding='utf-8' ) as file:
        strDataFile1 = file.read()
        strDataFile2 = strDataFile1[1:]
    
    resDataFile = Counter()
    if (inter):
        resDataFile = Counter(map(''.join, zip(strDataFile1, strDataFile2))) 
    else:
        resDataFile = Counter(map(''.join, zip(strDataFile1[::2], strDataFile2[::2])))    
    
    
    freqDict = dict(resDataFile)
    total = 0
    for k in freqDict.values():
        total += k
    
    for k in freqDict:
        freqDict[k] = float(freqDict[k]/total)
    return dict(sorted(freqDict.items(), key=lambda x: x[1], reverse=True))
```

```python
bigram_func('textWithoutSpaces', False)
```

Ентропія

```python
def count_entropy(ngram, entropy_dict):
    for k in entropy_dict:
        entropy_dict[k] = -entropy_dict[k] * math.log(entropy_dict[k], 2)
    total = 0
    for x in entropy_dict:
        total += entropy_dict[x]
    result = total/ngram
    return result
```

```python
count_entropy(1, mono_func('textWithSpaces'))
```

```python
count_entropy(1, mono_func('textWithoutSpaces'))
```

```python
count_entropy(2, bigram_func("textWithSpaces", True))
```

```python
count_entropy(2, bigram_func("textWithSpaces", False))
```

```python
count_entropy(2, bigram_func("textWithoutSpaces", True))
```

```python
count_entropy(2, bigram_func("textWithoutSpaces", False))
```

Надлишковість тексту

```python
def superfluity(x): #избыточность
    result = 1 - float(x / math.log(32,2))
    return result
    
```
